```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X,y = make_classification(n_features=2,n_samples=100,n_informative=1, hypercube=False,class_sep=3,
                         n_classes=2,n_redundant=0,n_clusters_per_class=1,random_state=42)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=41)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    w = np.ones(X.shape[1])
    lr = 0.1

    for i in range(500):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],w))
        w = w + lr*(y[j] - y_hat)*X[j]

    return w[0], w[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
X_input = np.linspace(-3,3,100)
y_input = m*X_input + b
```

```python
plt.plot(X_input,y_input,color='red')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```
