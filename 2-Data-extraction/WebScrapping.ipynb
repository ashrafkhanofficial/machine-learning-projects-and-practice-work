```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
response = requests.get(url)
```

```python
#response.content
```

```python
type(response.text)
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
cls = soup.find_all("div", class_="col-md-4 col-xl-4 col-lg-4")
```

```python
#cls
```

```python
len(cls)
```

```python
names = soup.find_all("a", class_="title")
```

```python
names
```

```python
name_list = []
for i in names:
    name_list.append(i.text.strip()) 
```

```python
name_list
```

```python
prices = soup.find_all("h4", class_="price float-end card-title pull-right")
```

```python
prices
```

```python
price_list = []
for i in prices:
    price_list.append(i.text.strip())
```

```python
price_list
```

```python
desc = soup.find_all("p", class_="description card-text")
```

```python
desc_list = []
for i in desc:
    desc_list.append(i.text.strip())
```

```python
dic = {"names":name_list, "Price":price_list, "description":desc_list}
```

```python
df = pd.DataFrame(dic).reset_index(drop=True)
```
