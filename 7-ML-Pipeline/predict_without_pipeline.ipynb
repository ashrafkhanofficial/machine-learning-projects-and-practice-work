```python
import pickle
import numpy as np
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl','rb'))
ohe_embark = pickle.load(open('models/ohe_embark.pkl','rb'))
clf = pickle.load(open('models/clf.pkl','rb'))
```

```python
user_input = np.array([2, 'male', 22, 'S'], dtype=object,).reshape(1,4)
```

```python
user_input.shape
```

```python
import pandas as pd
```

```python
sex = pd.DataFrame(user_input[:,1], columns=['sex'])
```

```python
user_input_sex = ohe_sex.transform(sex)
```

```python
user_input_sex
```

```python
embark = pd.DataFrame(user_input[:, 3], columns=['embark'])
```

```python
user_input_embark = ohe_embark.transform(embark)
```

```python
user_input_embark
```

```python
user_input_transformed = np.concatenate((user_input[:,[0,2]],user_input_sex,user_input_embark), axis=1)
```

```python
user_input_transformed.shape
```

```python
clf.predict(user_input_transformed)
```
